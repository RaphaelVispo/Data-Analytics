```python
import numpy as np 
import pandas as pd 
import os
```

**<h1>Cleaning the data</h1>**

importing, setting the rows and columns and removing redundant rows and columns

<h2> Importing the data to the program</h2>

```python
df = pd.read_csv("Higher-Education-Graduates-by-Discipline-Group-AY-2010-11-to-2018-19.csv")

df.head()
#df.tail()
```

<h2>Formatting the data </h2>

```python
#setting the colums
df2 = df.set_axis(df.iloc[1].values, axis=1, inplace=False)

#dropping the rows
df= df2.drop(df2.index[[0,1,-2,-1]])
df=df.reset_index(drop=True)
df
```

<h2>Converting the string to an int</h2>

```python
df[df.columns[1:]]=df[df.columns[1:]].replace(',', '', regex=True).astype(int)
df
```

**<h1>Data Analytics</h1>**


```python
changeddf=df[:-1].sort_values(' 2018-19 ', ascending=False)
changeddf.drop('Discipline Group', axis=1)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns


x = df.columns[1:-2]
y =np.array(changeddf.drop('Discipline Group', axis=1))
l=changeddf['Discipline Group'].values
plt.figure(figsize=(10,10))

a=0
for i in changeddf['Discipline Group'].values:
    plt.plot(
              x,
              y[a],
              label=i
            )
    a+=1
plt.yscale("log")

plt.title('All of the Discipline Groups Combined (log)')
plt.xlabel('Academic Year')
plt.ylabel('No. of Graduates')
plt.legend(bbox_to_anchor=(1,1))

'''handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(),bbox_to_anchor=(1,1))'''
plt.show()
```

The graph above is a logarithmic scale of the different fields with respect to the year. Looking at the graph, most of the trend are increasing. But, are interesting since there are decreasing quite dramatically. Lets explore these field furthur by solving the rate of change of the initial and final year


**<h2> Largest decrease in Number of Graduates </h2>**

Computing for the change from academic year 2009-10 to 2018-19.

Adding the rate of change column and the change column

```python
df[' Rate of ch ']=((df[' 2018-19 ']-df[' 2009-10 '])/df[' 2009-10 '])*100
df[' Change ']=(df[' 2018-19 ']-df[' 2009-10 '])
df.tail()
```

Arranging the in ascending form with respect to 'Change'

```python
ch_asc_df = df[:-1].sort_values(' Change ', ascending=True)
ch_asc_df.tail(3)
```

Looking at the data, it is the top 3 discipllines with decreasing graduates. **The highest decrease will be the 'Medical and Allied' discipline with a rate of change of -61.07% and a change of -71079 graduates**. This is unusual since most of the other fields have increasing numbers. There might be sociological reasons why it happened.

```python
fig, ax = plt.subplots(figsize=(12, 5))

sns.barplot(x=ch_asc_df.columns[1:-2],
        y=ch_asc_df.iloc[0:,1:-2].values[0]
            , palette='Blues_d')
plt.title(ch_asc_df['Discipline Group'].iloc[0])
plt.xlabel('Academic Year')
plt.ylabel('No. of Graduates')
plt.show()
```

The bar plot of the 'Medical and Allied'. As per the data, there was a sudden dip in the number of graduates from the Academic year 2009-10 to 2014-15.

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20, 4))

sns.barplot(ax= ax1,
            x=ch_asc_df.columns[1:-2],
            y=ch_asc_df.iloc[1:,1:-2].values[0],
            palette='YlOrRd')
ax1.set_title("{}".format(ch_asc_df['Discipline Group'].iloc[1]))
ax1.set_xlabel('Academic Year')
ax1.set_ylabel('No. of Graduates')
sns.barplot(ax= ax2,
            x=ch_asc_df.columns[1:-2],
            y=ch_asc_df.iloc[2:,1:-2].values[0],
            palette='YlOrBr')
ax2.set_title("{}".format(ch_asc_df['Discipline Group'].iloc[2]))
ax2.set_xlabel('Academic Year')
ax2.set_ylabel('No. of Graduates')
plt.show()
```

These are the other 2 Disciplines with decreasing numbers: 'Trade, Craft and Industrial' and 'General'. 

<h2> Largest Increases </h2>

```python
fig, ax = plt.subplots(2,2,figsize=(20, 10))
x=-1
c=['Greys_d', 'Greens_d', 'PuBuGn_d', 'Purples_d']


for a in range(2):
    for b in range(2):
        sns.barplot(ax=ax[a,b],
                x=ch_asc_df.columns[1:-2],
                y=ch_asc_df.iloc[x:,1:-2].values[0],
                palette=c[x])
        ax[a,b].set_title(ch_asc_df.iloc[x,0])
        ax[a,b].set_ylabel('No. of Graduates')
        ax[a,b].set_xlabel('Academic Year')
        x-=1
        
plt.show()
```

```python
fig, ax = plt.subplots(7,2,figsize=(19, 40))
x=3
c=sns.color_palette("Spectral", 14)

for a in range(7):
    for b in range(2):
        sns.barplot(ax=ax[a,b],
                x=ch_asc_df.columns[1:-2],
                y=ch_asc_df.iloc[x:,1:-2].values[0],
                color=c[x-3])
        ax[a,b].set_title(ch_asc_df.iloc[x,0])
        ax[a,b].set_ylabel('No. of Graduates')
        ax[a,b].set_xlabel('Academic Year')
        x+=1
        
plt.show()
```

The other disciplines bar charts are show above.

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20, 4))

sns.barplot(ax= ax1,
            x=list(ch_asc_df.sort_values(' 2009-10 ')['Discipline Group']),
            y=list(ch_asc_df.sort_values(' 2009-10 ')[' 2009-10 ']),
            palette='YlOrRd')
ax1.tick_params( labelrotation=90)

ax1.set_title("A.Y. 2009-10 vs Number of Graduates")
ax1.set_xlabel('Academic Year')
ax1.set_ylabel('Discipline Group')

sns.barplot(ax= ax2,
            x=list(ch_asc_df.sort_values(' 2018-19 ')['Discipline Group']),
            y=list(ch_asc_df.sort_values(' 2018-19 ')[' 2018-19 ']),
            palette='YlOrRd')
ax2.set_title("A.Y. 2018-19 vs Number of Graduates")
ax2.set_xlabel('Academic Year')
ax2.set_ylabel('Discipline Group')
ax2.tick_params( labelrotation=90)
plt.show()
```

Above is a change of the popularity of the degree based on the graduates between the Academic year 2009-10 and 2018-19:
* The top 5 Disciplines from  Academic year 2009-10: 'Business Administration', 'Medical and Allied', 'Education science and Teacher Training', 'IT-Related Discipline' and 'Engineering and Tech' respectively.
* The top 5 Disciplines from  Academic year 2018-19: 'Business Administration', 'Education science and Teacher Training', 'Engineering and Tech', 'IT-Related Discipline' and 'Other Disciplines', respectively.

```python
x = list(df.columns[1:-2])
y = np.array(changeddf.drop(['Discipline Group'], axis=1))
plt.figure(figsize=(10,10))

c=sns.color_palette("Set2", 21)

def plots(a, array):
    if a==21:
        return 0
    plt.bar(x,
            y[a],
            bottom=array,
            label=changeddf['Discipline Group'].iloc[a],
            color=c[a])
    
    if a==0:
        array=y[a]
    else:
        array+=y[a]
    plots(a+1, array)

plots(0,0)

plt.title('All of the Discipline Groups Combined')
plt.xlabel('Academic Year')
plt.ylabel('No. of Graduates')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

The change of the grand total of the graduates of the from Academic Year 2009-10 to 2018-19 is **315,245 with 65.49% increase.**
